```python
import re 

def preprocessing(text):
    
    #공백 두개 이상 있는 거 다 하나로 지우기 
    text = re.sub('  +', ' ', text)
    #\n (줄바꿈) 제거
    text = re.sub('\n', ' ', text)
    #특수기호 제거(underscore, 0-9는 삭제 안됨)
    text = re.sub('[^\w\s]', '', text)
    #underscore 제거
    text = re.sub('_', '', text)
    #'ㅋ+' 음절 제거
    text = re.sub('ㅋ+', '', text)
    #'ㅎ+' 음절 제거
    text = re.sub('ㅎ+', '', text)

    return text.strip()
```

```python
df['content'] = df['content'].apply(preprocessing)
df = df[df['content'].str.len() > 1] # 댓글 길이가 한 글자면 제거
```

# 각 어절에서 조합 가능한 모든 부분 글자 생성

```python
def generate_substrings(word):
    substrings = []
    for i in range(len(word)):
        for j in range(i + 1, len(word) + 1):
            substring = word[i:j]
            if len(substring) > 1:  # 길이가 1보다 큰 경우에만 추가
                substrings.append(substring)
    return substrings

def generate_all_substrings(text):
    words = text.split()  # 어절로 나누기
    all_substrings = []
    for word in words:
        substrings = generate_substrings(word)
        all_substrings.extend(substrings)
    return all_substrings
```

```python
df['substrs'] = df['content'].apply(generate_all_substrings)
```

```python
all_substrs = []
for i in df['substrs']:
    all_substrs.extend(i)
```
